```python
import matplotlib.pyplot as plt # importa a biblioteca matplotlib
import numpy as np # importa a biblioteca numpy
```

```python
# Gerandp os dados para plotar o gráfico
# Nesse caso, linspace gera valore espaçados de 10 em 10
Dados_x = np.linspace(0, 10, 10) # gera 10 valor de 10 em 10
Dados_y = Dados_x * Dados_x
```

```python
plt.figure(figsize=(10,8), facecolor='gray') # plt.figure
# cria a figura
# figsize tamanho da figura
Ax = plt.axes() # cria os eixos do gráfico
Ax.set_facecolor('white') # cor de fundo do gráfico
plt.plot(Dados_x,Dados_y,color='red',marker='o',linestyle='dashed',markeredgecolor='black')

# plota o gráfico em x e y
plt.xlabel('Eixo x',fontsize=16)
plt.ylabel('Eixo y',fontsize=16)
plt.title('Gráfico simples com Matplotlib',fontsize=16)
```

```python
# Subplot

# Gráfico 1
plt.rcParams.update({'font.size':15}) # função que permite ajustar a figura
plt.figure(figsize=(12,12)) # define um tamanho para a figura
linhas = 2 # cria 2 linhas
colunas = 2 # cia duas colunas
# Subplot para plotar as figuras
plt.subplot(linhas,colunas,1) # plota gráfico na primeira posição
plt.plot(Dados_x,'r--',label='Dados_x')
plt.legend(loc='lower right') # cria legenda para o gráfico
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.title('Gráfico 1')


# Gráfico 2
plt.subplot(linhas,colunas,2) # plota gráfico na primeira posição
plt.plot(Dados_x,Dados_y,'b--',label='Dados x \n e y') 
#plota gráfico alterando a cor da linha
plt.title('Gráfico 2') # título do gráfico
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.legend

# Gráfico 3
plt.subplot(linhas,colunas,3) # plota gráfico na primeira posição
plt.plot(Dados_x * Dados_x,'g-',label='Dados x^2') 
#plota gráfico alterando a cor da linha
plt.title('Gráfico 3') # título do gráfico
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.legend

# Gráfico 4
plt.subplot(linhas,colunas,4) # plota gráfico na primeira posição
plt.plot(Dados_y, 'c--',label='Dados_y') 
#plota gráfico alterando a cor da linha
plt.title('Gráfico 4') # título do gráfico
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.legend()
plt.savefig('Gráficos aula-3.png') # serve para salvar a figura(gráfico)
plt.show() # remove  warning

```
