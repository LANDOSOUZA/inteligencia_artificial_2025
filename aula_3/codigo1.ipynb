```python
# Importando as bibliotecas

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Realizar a leitura do meu arquivo csv
df = pd.read_csv('Base de Dados/monitoramento_tempo.csv')
```

```python
# Exibindo as primeiras linhas do dataframe
df.head() # exibe as 4 primeiras linha do dataframe
```

```python
# Importa a biblioteca datetime
import datetime # Biblioteca que permite a transformação da data que era um objeto em datetime

df['data'] = pd.to_datetime(df['data']) # convete a data de object para datetime
plt.figure(figsize=(15, 8))
plt.plot(df['data'], df['temperatura'])
plt.xlabel('Data e hora')
plt.ylabel('Temperatura')
```

```python
fig = plt.figure(figsize=(10, 8)) # cria uma variável fig para armazenar o tamanho da figura
eixo = fig.add_axes([0, 0, 1, 1]) # define o tamanho dos eixos do gráfico
eixo.plot(df['data'], df['temperatura']/10, color='red', lw=0.2, linestyle = '--', marker = '')
eixo.set_title('Temperatura no momento', fontsize=25)
eixo.legend(['temperatura'], loc='lower right', fontsize=15)
eixo.set_xlabel('Ano', fontsize=20)
eixo.set_ylabel('Temperatura', fontsize=20)
```

```python
# Selecionado o período da base de dados

fig = plt.figure(figsize=(15, 8)) # cria uma variável fig para armazenar o tamanho da figura
plt.rcParams['font.size']='20'
eixo = fig.add_axes([0, 0, 1, 1]) # define o tamanho dos eixos do gráfico
eixo.plot(df['data'], df['temperatura'],color='b', lw=0.5, linestyle = '--')
eixo.set_xlim(datetime.datetime(2015,1,1),datetime.datetime(2015,12,31)) # Xlin define o valor inicial e o valor final
eixo.set_title('Temperatura - Ano 2015', fontsize=25)
eixo.legend(['temperatura'], fontsize=20)
eixo.set_xlabel('Ano', fontsize=20)
plt.xticks(rotation=45,fontsize=15,ha='right') # Ajusta o espaçamento
plt.show
```
