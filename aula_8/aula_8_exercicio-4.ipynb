```python
# Em uma planta de produção industrial, você deseja classificar as máquinas com base em 
# sua eficiência energética. As características incluem o consumo de energia, o tempo de 
# operação e o tipo de energia utilizada (por exemplo, elétrica ou gás). Use uma árvore de 
# decisão para classificar as áquinas e, "Eficientes " e "Ineficientes" em termos de usuo 
# de energia.

# 'Consumo de Energia (KW/h)': [1000, 3000, 1500, 2500, 1200, 3500],
# 'Tempo de Operação (hora)': [200, 50, 300, 400, 250, 600],
# 'Tipo de Energia': ['Elétrica', 'Gás', 'Elétrica', 'Gás', 'Eletrica', 'Gás'],
# 'Classificação': ['Eficiente', 'Ineficiente', 'Eficiente', 'Ineficiente', 'Eficiente', 'Ineficiente']
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
# Dados simulados
dados = pd.DataFrame({
    'Consumo de Energia (KW/h)': [1000, 3000, 1500, 2500, 1200, 3500],
    'Tempo de Operação (hora)': [200, 50, 300, 400, 250, 600],
    'Tipo de Energia': ['Elétrica', 'Gás', 'Elétrica', 'Gás', 'Eletrica', 'Gás'],
    'Classificação': ['Eficiente', 'Ineficiente', 'Eficiente', 'Ineficiente', 'Eficiente', 'Ineficiente']
})

# Corrigindo possíveis inconsistências de grafia
dados['Tipo de Energia'] = dados['Tipo de Energia'].str.lower().str.capitalize()

# Codificando variáveis categóricas
X = pd.get_dummies(dados.drop('Classificação', axis=1))
y = dados['Classificação']

# Treinando o modelo
modelo = DecisionTreeClassifier()
modelo.fit(X, y)

# Visualizando a árvore
plt.figure(figsize=(12, 8))
tree.plot_tree(modelo, filled=True, feature_names=X.columns, class_names=modelo.classes_)
plt.title("Classificação de Máquinas por Eficiência Energética")
plt.show()
```
