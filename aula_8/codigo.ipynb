```python
# importação das bibliotecas
import numpy as np # importa a biblioteca numpy
import matplotlib.pyplot as plt # importação da biblioteca matplotlib
import seaborn as sns # imporrtação da biblioteca seaborn
from sklearn.model_selection import train_test_split # importação da biblioteca para divir os dados em treino e teste
from sklearn.tree import DecisionTreeClassifier # importação a árvore de decisão
from sklearn import tree # importação do elemento tree para plotar a árvore
```

```python
# !pip install seaborn - instalação da biblioteca
```

```python
# Criação do conjunto de dados
notas = np.array([8, 6, 5, 9, 4, 3, 7, 2, 1, 5, 6, 9, 8])
classes = np.array(['bom', 'ruim', 'ruim,', 'bom', 'péssimo', 'péssimo', 'bom', 'péssimo', 'péssimo', 'ruim', 'ruim', 'bom', 'bom'])
```

```python
# Divisão da base de dados em treino e teste
notas_treino,notas_teste,classes_treino,classes_teste=train_test_split(notas.reshape(-1,1),
                                                                       classes.reshape(-1,1),
                                                                       test_size=0.2, # 20% dos dados para treino
                                                                       random_state=42 ) # permite a resprodutibilidade dos testes
```

```python
# Criação do modelo de Machine Learning Árvore de Decisão
modelo = DecisionTreeClassifier() # Criação do mdelo
modelo.fit(notas_treino,classes_treino) # Treinamento do modelo
```

```python
# Teste do modelo
print(modelo.predict(notas_teste))
print(notas_teste)
```

```python
modelo.predict(np.array([4]).reshape(-1,1)) # Teste de uma nota individualmente
```

```python
modelo.predict(np.array([7.5]).reshape(-1,1)) # Teste de uma nota individualmente
```

```python
modelo.predict(np.array([6, 5, 9, 7]).reshape(-1,1)) # Teste de uma nota individualmente
```

```python
# Visualização do modelo de árvore de decisão
fig = plt.figure(figsize=(15,4)) # Criação de uma figura do tamanho 10x8
# tree.plot = plota a árvore de decisão
# modelo é o aprendizado supevisionado
tree.plot_tree(modelo,feature_names=notas.tolist(),class_names=classes.tolist(),filled=True)
plt.title('Árvore de decisão para classificar uma bebida \n de acordo com a nota',fontsize=18)
```
