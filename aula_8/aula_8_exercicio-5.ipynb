```python
# Suponha que você é responsável por determinar quando realizar a manutenção preventiva 
# em diferentes máquinas em uma fábrica de processamento de alimentos. As características 
# incluem a idade das máquinas, o número de horas de operação desde a última manutenção e 
# o histórico de falhas. Use uma árvore de decisão para classificar as máquinas em 
# "Necessidade de Manutenção" e Não Necessidade de Manutenção".

# 'Idade (ano)': [2, 6, 1, 4, 3, 5],
# 'Horas Desde a Última Manutenção': [50, 200, 20, 150, 100, 180],
# 'Histórico de Falhas': ['Nenhuma', 2, 'Nenhuma', 1, 3, 2],
# 'Classificação': ['Necessita, Necessita, Necessita, Não Necessita, Necessita, Não Necessita']
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
# Dados simulados
dados = pd.DataFrame({
    'Idade (ano)': [2, 6, 1, 4, 3, 5],
    'Horas Desde a Última Manutenção': [50, 200, 20, 150, 100, 180],
    'Histórico de Falhas': ['Nenhuma', 2, 'Nenhuma', 1, 3, 2],
    'Classificação': ['Necessita', 'Necessita', 'Necessita', 'Não Necessita', 'Necessita', 'Não Necessita']
})

# Padronizando valores categóricos
dados['Histórico de Falhas'] = dados['Histórico de Falhas'].replace('Nenhuma', 0)

# Separando variáveis preditoras e alvo
X = dados.drop('Classificação', axis=1)
y = dados['Classificação']

# Treinando o modelo
modelo = DecisionTreeClassifier()
modelo.fit(X, y)

nova_maquina = pd.DataFrame({
    'Idade (ano)': [3],
    'Horas Desde a Última Manutenção': [120],
    'Histórico de Falhas': [1]
})

print("Classificação prevista:", modelo.predict(nova_maquina)[0])


# Visualizando a árvore
plt.figure(figsize=(12, 8))
tree.plot_tree(modelo, filled=True, feature_names=X.columns, class_names=modelo.classes_)
plt.title("Classificação de Máquinas para Manutenção Preventiva")
plt.show()
```
