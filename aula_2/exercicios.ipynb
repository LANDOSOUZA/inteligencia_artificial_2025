```python
# 1) Plote o gráfico umidade x ano, colocando título 
#    na imagem, título dos eixos, alterando o 
#    tamanho da imagem e da fonte

import pandas as pd
import matplotlib.pyplot as plt

# Carrega o CSV
df = pd.read_csv('monitoramento_tempo.csv')

# Converte a coluna 'data' para datetime
df['data'] = pd.to_datetime(df['data'])

# Extrai o ano da data
df['ano'] = df['data'].dt.year

# Agrupa por ano e calcula a média da umidade
umidade_por_ano = df.groupby('ano')['humidade'].mean()

# Cria o gráfico
plt.figure(figsize=(10, 6))  # Tamanho da imagem

plt.plot(umidade_por_ano.index, umidade_por_ano.values, marker='o', color='blue')

# Títulos e rótulos com fonte ajustada
plt.title('Variação da Umidade por Ano', fontsize=16)
plt.xlabel('Ano', fontsize=14)
plt.ylabel('Umidade Média (%)', fontsize=14)

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#  3) Realizar anotações no gráfico de umidade

import pandas as pd
import matplotlib.pyplot as plt

# Carrega os dados
df = pd.read_csv('monitoramento_tempo.csv')
df['data'] = pd.to_datetime(df['data'])
df['ano'] = df['data'].dt.year

# Agrupa por ano
umidade_por_ano = df.groupby('ano')['humidade'].mean()

# Identifica os anos com maior e menor umidade
ano_max = umidade_por_ano.idxmax()
umidade_max = umidade_por_ano.max()

ano_min = umidade_por_ano.idxmin()
umidade_min = umidade_por_ano.min()

# Cria o gráfico
plt.figure(figsize=(10, 6))
plt.plot(umidade_por_ano.index, umidade_por_ano.values, marker='o', color='blue')

# Títulos e eixos
plt.title('Variação da Umidade por Ano', fontsize=16)
plt.xlabel('Ano', fontsize=14)
plt.ylabel('Umidade Média (%)', fontsize=14)

# Anotações
plt.annotate(f'Máxima: {umidade_max:.1f}%', 
             xy=(ano_max, umidade_max), 
             xytext=(ano_max, umidade_max + 5),
             arrowprops=dict(facecolor='green', shrink=0.05),
             fontsize=12, color='green')

plt.annotate(f'Mínima: {umidade_min:.1f}%', 
             xy=(ano_min, umidade_min), 
             xytext=(ano_min, umidade_min - 10),
             arrowprops=dict(facecolor='red', shrink=0.05),
             fontsize=12, color='red')

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 4) Realizar anotações no gráfico de direção do vento

import pandas as pd
import matplotlib.pyplot as plt

# Carrega os dados
df = pd.read_csv('monitoramento_tempo.csv')
df['data'] = pd.to_datetime(df['data'])
df['ano'] = df['data'].dt.year

# Agrupa por ano e calcula a média da direção do vento
direcao_por_ano = df.groupby('ano')['direção do vento'].mean()

# Identifica os extremos
ano_max = direcao_por_ano.idxmax()
direcao_max = direcao_por_ano.max()

ano_min = direcao_por_ano.idxmin()
direcao_min = direcao_por_ano.min()

# Cria o gráfico
plt.figure(figsize=(10, 6))
plt.plot(direcao_por_ano.index, direcao_por_ano.values, marker='o', color='purple')

# Títulos e eixos
plt.title('Direção Média do Vento por Ano', fontsize=16)
plt.xlabel('Ano', fontsize=14)
plt.ylabel('Direção do Vento (graus)', fontsize=14)

# Anotações
plt.annotate(f'Máxima: {direcao_max:.1f}°',
             xy=(ano_max, direcao_max),
             xytext=(ano_max, direcao_max + 20),
             arrowprops=dict(facecolor='green', shrink=0.05),
             fontsize=12, color='green')

plt.annotate(f'Mínima: {direcao_min:.1f}°',
             xy=(ano_min, direcao_min),
             xytext=(ano_min, direcao_min - 30),
             arrowprops=dict(facecolor='red', shrink=0.05),
             fontsize=12, color='red')

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 6) Plotar boxplot para a umidade

import pandas as pd
import matplotlib.pyplot as plt

# Carrega os dados
df = pd.read_csv('monitoramento_tempo.csv')

# Cria o boxplot
plt.figure(figsize=(8, 6))  # Tamanho da imagem
plt.boxplot(df['humidade'], patch_artist=True, boxprops=dict(facecolor='skyblue'))

# Títulos e rótulos
plt.title('Distribuição da Umidade', fontsize=16)
plt.ylabel('Umidade (%)', fontsize=14)

plt.grid(True)
plt.tight_layout()
plt.show()
```
