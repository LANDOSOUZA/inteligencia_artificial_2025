```python
# 1) Plote o gráfico umidade x ano, colocando título 
#    na imagem, título dos eixos, alterando o 
#    tamanho da imagem e da fonte

import pandas as pd
import matplotlib.pyplot as plt

# Carrega o CSV
df = pd.read_csv('monitoramento_tempo.csv')

# Converte a coluna 'data' para datetime
df['data'] = pd.to_datetime(df['data'])

# Extrai o ano da data
df['ano'] = df['data'].dt.year

# Agrupa por ano e calcula a média da umidade
umidade_por_ano = df.groupby('ano')['humidade'].mean()

# Cria o gráfico
plt.figure(figsize=(10, 6))  # Tamanho da imagem

plt.plot(umidade_por_ano.index, umidade_por_ano.values, marker='o', color='blue')

# Títulos e rótulos com fonte ajustada
plt.title('Variação da Umidade por Ano', fontsize=16)
plt.xlabel('Ano', fontsize=14)
plt.ylabel('Umidade Média (%)', fontsize=14)

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 2) Plote o gráfico velocidade do ventox ano, 
#    colocando título na imagem, título dos eixos, 
#    alterando o tamanho da imagem e da fonte

import pandas as pd
import matplotlib.pyplot as plt

# Carrega o CSV
df = pd.read_csv('monitoramento_tempo.csv')

# Converte a coluna 'data' para datetime
df['data'] = pd.to_datetime(df['data'])

# Extrai o ano
df['ano'] = df['data'].dt.year

# Agrupa por ano e calcula a média da velocidade do vento
vento_por_ano = df.groupby('ano')['velocidade do vento'].mean()

# Cria o gráfico
plt.figure(figsize=(10, 6))  # Tamanho da imagem

plt.plot(vento_por_ano.index, vento_por_ano.values, marker='o', color='green')

# Títulos e rótulos com fonte ajustada
plt.title('Velocidade Média do Vento por Ano', fontsize=16)
plt.xlabel('Ano', fontsize=14)
plt.ylabel('Velocidade do Vento (m/s)', fontsize=14)

plt.grid(True)
plt.tight_layout()
plt.show()

```
